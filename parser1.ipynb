```python
from pdf2image import convert_from_path
import easyocr
import numpy as np
import PIL
from PIL import ImageDraw
import spacy
```

```python
reader = easyocr.Reader(['en'])
```

```python
images = convert_from_path('E:/Coding/bigsparkCVssample/YasseenAhmanache.pdf')

from IPython.display import display
from PIL import Image


display(images[1])
```

```python
bounds = reader.readtext(np.array(images[0]),paragraph='True',slope_ths=100)

def draw_boxes(image, bounds, color='red', width=2):
 draw = ImageDraw.Draw(image)
 for bound in bounds:
  p0, p1, p2, p3 = bound[0]
  draw.line([*p1, *p1, *p2, *p3, *p0], fill=color, width=width)
 return image

draw_boxes(images[0], bounds)
```
